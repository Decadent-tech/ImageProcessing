# Ways to show relationship within two sets of values to see is there in corelation

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

**Basic Plotting**

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
plt.scatter(x,y)
plt.show()
```

**Changing size and color and mark it as X also add an edge linewidth **

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
plt.scatter(x,y,s=100,c='green',marker='X')
plt.show()
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
plt.scatter(x,y,s=100,c='green',edgecolors='black',linewidths=1,alpha=0.75)
plt.show()
```

# Colors represent here

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y,s=100,c=colors,edgecolors='black',linewidths=1,alpha=0.75)
```

# Matplotlib Plot with Color Mapping using Matplotlib.set_cmap()

```python
plt.scatter(x,y,s=100,c=colors,cmap='Greens',edgecolors='black',linewidths=1,alpha=0.75)
```

# cbar - this would give a status bar range of colors

```python

plt.scatter(x,y,s=100,c=colors,cmap='Greens',edgecolors='black',linewidths=1,alpha=0.75)
cbar=plt.colorbar()
cbar.set_label('Satisfaction')
```

# We can add sizes as well which acts as visual representation for satisfaction level

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
plt.scatter(x,y,s=sizes,c=colors,cmap='Greens',edgecolors='black',linewidths=1,alpha=0.75)
cbar=plt.colorbar()
cbar.set_label('Satisfaction')
```

# Now focusing on real data which has data of top 200  videos

```python

data = pd.read_csv('2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

# **We can see 1 outlier which is a Taylor song with huge likes **

```python
plt.scatter(view_count,likes,
            edgecolors='black',linewidths=1,alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

# Now adding logarithmic scale

```python
plt.scatter(view_count,likes,
            edgecolors='black',linewidths=1,alpha=0.75)


plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```

# cmap used again to differentiate colorbar ration ,

*Colors represent like - dislike ration , green less performed , bright yellow performed more medium were more around mild green , some dark ones are *

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
