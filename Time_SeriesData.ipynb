```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates#date formatter class we would use
```

```python
plt.style.use('seaborn')
```

```python
dates = [
    datetime(2024, 8, 24),
    datetime(2024, 8, 25),
    datetime(2024, 8, 26),
    datetime(2024, 8, 27),
    datetime(2024, 8, 28),
    datetime(2024, 8, 29),
    datetime(2024, 8, 30)
]
```

```python
y = [0, 1, 3, 4, 6, 5, 7]#seven random values
```

# **Dates Value plotted in **

```python
plt.plot_date(dates, y)
```

# **Mark in 1 line **

```python
plt.plot_date(dates, y, linestyle='solid')
```

# **Date ticklabels often overlap, so it is useful to rotate them and right align them**

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()
plt.show()
```

# **Date Formatting as Month name in date **

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()#get current figure
date_format=mpl_dates.DateFormatter('%b, %d %Y')
#set format for x-axis
plt.gca().xaxis.set_major_formatter(date_format)#gca is get current axis
plt.show()
```

```python
data = pd.read_csv('data (2).csv')
price_date = data['Date']
price_close = data['Close']
```

# Dates that are out of oder we need to format them convert date to datetime , also sort them in oder and make inplace

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()#get current figure
date_format=mpl_dates.DateFormatter('%b, %d %Y')
#set format for x-axis
plt.gca().xaxis.set_major_formatter(date_format)#gca is get current axis

plt.title('Bitcoin Prices')
plt.xlabel('7 -Days -In')
plt.ylabel('Closing Price')

plt.show()
```
