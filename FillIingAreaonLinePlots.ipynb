```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_excel('Book1-FillBetweens.xlsx')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

**Fill in line  plots with Python and All developers salary  **

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**Fill in the entire area under python line  plots with **




```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287
plt.fill_between(ages, py_salaries)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**Filling in line plots with alpha it creates conditional pressure **

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

# overall_median = 57287
plt.fill_between(ages, py_salaries,alpha=0.25)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**Python Plot crossed overall median salary **

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median,alpha=0.25)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**Our Python Plot is over our median salary **

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median,
                 where=(py_salaries>overall_median),
                 interpolate=True ,alpha=0.25)
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

**Now we have our plots filling in between area of fixed plots and fixed value **

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median,
                 where=(py_salaries>overall_median),
                 interpolate=True ,alpha=0.25)
plt.fill_between(ages, py_salaries,overall_median,
                 where=(py_salaries<=overall_median),
                 interpolate=True ,alpha=0.25,color='red')
plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
