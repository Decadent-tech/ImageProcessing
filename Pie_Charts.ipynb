```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

# Basic

```python
slices=[120,80,30,20]
labels=['Sixty','Forty','Other','Other2']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']

plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

**Explode**

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

**Shadow Added**

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

**Adding a 90deg**

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,startangle=90,wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```

**Adding Percentage**

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,
        startangle=90,autopct='%1.1f%%',
        wedgeprops={'edgecolor':'black'})
plt.title('Pie')
plt.tight_layout()
plt.show()
```
