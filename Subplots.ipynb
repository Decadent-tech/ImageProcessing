# Stateful - we have figures and axis on this plot

```
figure - containers that hold multiple plot
and axis - are actual plot

Figure can have multiple plots
plt=Plot()
plt.gcf()-figure
plt.gca()-axis
```



````python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
````

````python
plt.style.use('seaborn')
````

````python
data = pd.read_csv('data (4).csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
````

# Sub-Plot method

```
**instead of having all data in one plot we want multiple plots add more axis we need sub plots **
```



````python
fig, ax = plt.subplots()

data = pd.read_csv('data (4).csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

````

# Breaking Subplots

````python
fig, ax = plt.subplots()
print(ax)#single axis
````

````python
fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax)#2 axes
````

````python
fig, ax = plt.subplots(nrows=2, ncols=2)
print(ax)#4 axes
````

# Unpack the Plots

````python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=2)
print(ax1)#4 axes
print(ax2)
````

# Code for Developers and Python Dev SubPlots

````python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

````
